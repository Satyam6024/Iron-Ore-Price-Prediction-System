# 04 Feature Selection

This notebook shows how the full engineered feature set is reduced before training.

Current approach:
- rank features using `random_forest_importance`
- keep top `k` features
- force-include core market features such as `Open`, `High`, `Low`, `Vol.`, and `Change %`

# 07 MLflow Register

This notebook registers the trained model into MLflow Model Registry.

Note: running the registration cell creates a new model version in the configured registry backend.


```python
from pathlib import Path
import json
import os
import subprocess
import sys
import yaml

PROJECT_ROOT = Path.cwd().resolve()
if not (PROJECT_ROOT / "params.yaml").exists():
    PROJECT_ROOT = PROJECT_ROOT.parent
os.chdir(PROJECT_ROOT)

def read_json(path: str):
    return json.loads(Path(path).read_text(encoding="utf-8"))

params = yaml.safe_load(Path("params.yaml").read_text(encoding="utf-8"))
```

```python
{
    "mlflow": params["mlflow"],
    "registry": params["registry"],
}
```

```python
subprocess.run([sys.executable, "dvc_pipeline/src/mlflow_register_model.py"], check=True)
```
